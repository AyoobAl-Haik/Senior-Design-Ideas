```python
import numpy as np

# Example: Create a 5x5 array of zeros
scan_area = np.zeros((5, 5), dtype=int)

# Set one value greater than 0 to test break condition
scan_area[2, 3] = 1  # This will trigger the break

# Raster scan with early break
for row_index in range(scan_area.shape[0]):
    for col_index in range(scan_area.shape[1]):
        if scan_area[row_index, col_index] > 0:
            print(f"Value > 0 found at ({row_index}, {col_index}):", scan_area[row_index, col_index])
            break
    else:
        # Only executes if the inner loop was not broken
        continue
    # Breaks outer loop if inner loop was broken
    break
```
